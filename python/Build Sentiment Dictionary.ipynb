This notebook contains steps to construct a sentiment dictionary from the corpus of Yelp reviews.

# Outline

Outline
1. Tokenize text using Potts sentiment tokenizer
2. Label text with part-of-speech tags using nltk (collapsing to n,a,v,r)
3. Lemmatize words using ntk.stem.wordnet.WordNetLemmatizer 
4. Drop words with less than N mentions in the corpus
5. Plot some words and their relative frequencies vs. star rating
6. Perform regression (linear or logistic based on plots above) for each word
7. Create dictionary in which keys are results of nltk.stem.wordnet.WordNetLemmatizer and values are the slopes of the regression (maybe normalized to range [-1,1]).

```python
import pandas as pd
import numpy as np
import nltk
import external.potts_tokenizer as ptk  # sentiment tokenizer
import sentiment
```

```python
# Read mexican restaurant review data
reviews = pd.read_pickle('../data/pandas/review_mexican_train.pkl')
```

### 1. Tokenize

```python
# Split reviews into sentences
sentences = []
#sentences = pd.DataFrame(columns=['review_id','sentence_id','text','stars'])
total_count = 0
for rev_id in reviews.review_id:
    #if total_count % 1000==0: print total_count
    thisreview = reviews[reviews.review_id==rev_id]
    sents = nltk.sent_tokenize(thisreview.text.values[0]) # nltk sentence tokenize
    thissent_count = 0
    for sent in sents:
        sentid = str(thissent_count).zfill(5)
        thisseries = pd.Series([thisreview.review_id.values[0],sentid,sent,thisreview.stars.values[0]])
        sentences.append(thisseries)
        thissent_count += 1
        total_count += 1

# Convert to dataframe and save sentences dataframe to file
dataframe = pd.concat(sentences,axis=1).T
dataframe.columns = ['review_id','sentence_id','text','stars']
dataframe.to_pickle('../data/pandas/sentences_mexican_train.pkl')
```

```python
# Read sentences dataframe from file
sentences = pd.read_pickle('../data/pandas/sentences_mexican_train.pkl')
sentences
```

```python
# TEMP MAKE BAGS
bags = []
for nstar in sentences.stars.unique():
    print "processing stars = {}".format(nstar)
    thisdata = sentences[sentences['stars']==nstar]
    thisbag = sentiment.get_bag(thisdata)
    thisbag.name = str(int(nstar))
    bags.append( thisbag )
    
dataframe = pd.concat(bags,axis=1)
dataframe.to_pickle('../data/pandas/word_count_mexican_train.pkl')
```
