# Outline

This notebok evaluates FoodFindr sentence scores compared to restaurants' Yelp scores.
1. Load pandas dataframe of sentences from pickled file
2. Calculate average FoodFindr sentence score for each restaurant
3. Plot FoodFindr sentence scores (y-axis) vs. Yelp rating (x-axis) for each restaurant
4. Interpretation of the plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Load Sentences

```python
sentences = pd.read_pickle('../data/pandas/sentences_lemmas_scored_mexican.pkl')
reviews = pd.read_pickle('../data/pandas/review_mexican.pkl')
```

```python
# Join business ID to sentences dataframe for averaging by business
sentences = pd.merge(sentences,reviews[['business_id','review_id']],left_on="review_id", right_on="review_id")
sentences.stars = sentences.stars.astype(int)
```

# 2. Average FoodFindr score by restaurant

```python
# Sentence dataframe grouped by business ID (individual restuarnts)
grouped_sentences = sentences.groupby('business_id')
grouped = pd.DataFrame()
grouped['FF_score'] = grouped_sentences['FF_score'].mean()
grouped['Yelp_score'] = grouped_sentences['stars'].mean()
grouped['Nsentences'] = grouped_sentences['review_id'].count()
grouped['Nreviews'] = grouped_sentences['review_id'].nunique()
```

# 3. Plot FoodFindr restaurant averages vs. Yelp ratings

```python
for nreview in [1,10]:

    # Filter restaurants with few reviews
    plotdata = grouped[grouped['Nreviews']>=nreview]

    # Set up figure
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)

    # Plot FoodFindr vs. Yelp average restaurant scores
    ax.plot(np.arange(7),np.arange(7),'--k',lw=2,label='1:1')
    ax.plot(plotdata['Yelp_score'],plotdata['FF_score'],'or',ms=12,alpha=0.3,label='data')

    ax.set_xlim([0.5,5.5])
    ax.set_ylim([0.5,5.5])
    ax.set_xlabel('Yelp Score')
    ax.set_ylabel('FoodFindr Score')
    ax.legend(loc='best',numpoints=1)
    ax.set_title('Restaurants with >={} reviews'.format(nreview))
    plt.rcParams.update({'font.size': 22})
```

# 4. Interpretation

#### Quick Summary:

1. FoodFindr average restaurant scores generally correlates well with Yelp ratings. 
2. FoodFindr is biased high at the low end of ratings.
3. The correlation is noiser, but still holds for restaurants with less than 10 reviews.

#### Longer Interpretation:

1. Removing restaurants with few reviews reduces noise in the comparison between FoodFindr and Yelp restaurant scores. I'll focus on the lower plot to avoid interpreting noise.
2. The most obvious feature here is the departure from the 1:1 line at the low end of Yelp scores. I have two hypotheses that could explain this feature:
    - People are harsher with their numerical evaluation (1-5 star label) than they are with their sentences. This is certainly plausible, and I have seen 1-star reviews that say things like, "We waited a long time for our table. The food was pretty good. We were treated so rudely that we'll never come back."
    - FoodFindr is biased and rates crappy sentences too highly. 
        - 1) Because 1-star is at the end of the spectrum, any error in classifying 1-star sentences will represent a high bias. However, we would therefore expect the same problem with 5-star reviews. The problem is much less pronounced for highly rated restaurants.
        - 2) The vocabulary that people use when writing 2-3 star reviews is not as distinct as when writing the 1, 4, and 5 star reviews. Therefore, fewer key words indicate the existence of the 2-3 star reviews, and FoodFindr has trouble classifying those sentences.
3. At the highest end of the Yelp scores, FoodFindr slightly underpredicts the average restaurant score. One of the explanations above could explain this small feature, but because the departure from the 1:1 line here is small, another phenomenon could be the cause. I will speculate that rounding errors in the Yelp reviews could be responsible for this discrepancy. By this, I mean that Yelp users are forced to select a whole number between 1-5, even though their true feelings probably follow a continuous distribution between 1-5. A user that feels "4.6" about a restaurant will have to round up his/her review to a 5, incurring a rounding error. From the sentences themselves, we may more precisely evaluate the customer sentiment. Ok, this gives FoodFindr the benefit of the doubt, assuming that it accurately classifies each sentence.
