# 1. Test Naive Bayes Algorithm

```python
# Restore review data to train and test model
dataframe = pd.read_pickle('../data/pandas/review_lemmas_mexican.pkl')
X, y, index = classification.build_training_input(dataframe)
```

### Validation on Small Reviews Only

```python
def count_word(string):
    """
    Count number of sentences in a given string.
    """
    import nltk
    sents = nltk.sent_tokenize(string)
    words = []
    for s in sents:
        words.extend(nltk.word_tokenize(s))
        
    return len(words)

wordlen = dataframe['text'].map(count_word)
```

```python
def count_sent(string):
    """
    Count number of sentences in a given string.
    """
    import nltk
    sents = nltk.sent_tokenize(string)
    return len(sents)

senlen = dataframe['text'].map(count_sent)
```

```python
dataframe = pd.read_pickle('../data/pandas/review_lemmas_mexican.pkl')

# Add sentence and word lenghts to the pandas dataframe
dataframe['senlen'] = senlen
dataframe['wordlen'] = wordlen

# Add my review scores to the pandas dataframe
myinds = np.append(index_train,index_test)
myscore = pd.Series(np.append(results.predict(Xtrain),results.predict(Xtest)),index=myinds)
dataframe['myscore'] = myscore

# Add indicator of whether it was test or training data
dataframe['traintest'] = '0'
dataframe.loc[index_train,'traintest'] = 'train'
dataframe.loc[index_test,'traintest'] = 'test'
```

```python
# Figure out how to trim results to include only reviews with small number of sentences
testdf = dataframe[dataframe['traintest']=='test']

count = []
score = []
start,end = 1,16
for k in range(start,end):
    #tempdf = testdf[(testdf['senlen']<=k) & ((testdf['stars']!=1) & (testdf['stars']!=5))]
    tempdf = testdf[((testdf['senlen']==k) | (testdf['senlen']==k+1))]# | (testdf['senlen']==k+2))]# & 
                    #((testdf['stars']==1) | (testdf['stars']==5))]
    ysmall = tempdf['myscore']
    ytrue  = tempdf['stars']

    frac2d = np.zeros((5,5))
    num = 0
    denom = 0
    #for ii in [2,3,4]:
    for ii in [1,2,3,4,5]:
        thesevals = ysmall[ytrue==ii]
        thesevals = np.histogram(thesevals.astype(np.int32),bins=[0.5,1.5,2.5,3.5,4.5,5.5])
        frac2d[:,ii-1] = thesevals[0]/float(sum(thesevals[0]))
        num = num + thesevals[0][ii-1]
        denom = denom + float(sum(thesevals[0]))
    score.append(num/denom)
    count.append(len(ysmall))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
xaxis = np.arange(start,end)
yticks1 = [0,500,1000,1500,2000]
yticks2 = [0.5,0.52,0.54,0.56,0.58,0.6]
xticks = [0,5,10,15]
ln1 = ax1.plot(xaxis,count,'-b',lw=3,label='Count')
ax2 = ax1.twinx()
ln2 = ax2.plot(xaxis,score,'-k',lw=3,label='Accuracy')

ax1.set_yticks(yticks1)
ax1.set_yticklabels([str(y) for y in yticks1],fontsize=20,color='b')
ax1.set_xticks(xticks)
ax1.set_xticklabels([str(x) for x in xticks],fontsize=20)
ax1.set_xlabel('# Sentences/Review',fontsize=28)
#ax2.set_ylabel('Count', fontsize=28)
ax2.set_yticks(yticks2)
ax2.set_yticklabels([str(y) for y in yticks2],fontsize=20)
#ax1.set_ylabel('Mean Accuracy', fontsize=28)

lns = ln1+ln2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0,fontsize=20)

plt.savefig('img/accurace_vs_ReviewLength.png')
```
